Noob


```sql
--Count employees in each department.
select department,count(*) as total_emp from employees group by department
```

```sql
--Count employees in each city.
select city,count(*) as total_people from employees group by city
```

```sql
--Count employees per job title.
select job_title,count(*)as total_emp from employees group by job_title
```

```sql
--Sum of salaries per department.
select department,sum(salary)as totalsalary from employees group by department
```

```sql
--Average salary per department.
select department,avg(salary)as avg_salary from employees group by department
```

```sql
--Count employees with each status (active, inactive, etc.).
select status,count(*)as status from employees group by status
```

```sql
--Count employees per department per city.
select department,city,count(*)as total_emp from employees group by department,city
```

```sql
--Find the minimum salary in each department.
select department,min(salary)as low_salary from employees group by department
```

```sql
--Find the maximum salary in each city.
select city,round(max(salary))as high_salary from employees group by city
```

```sql
--Count employees hired on each date.
select hire_date,count(*) as hired_date from employees group by hire_date
```

Intermediate

```sql
--Count employees hired per year.
select year(hire_date),count(*) as hire_year from employees group by year(hire_date)
```

```sql
--Count employees hired per month.
select month(hire_date)as hire_month,count(*) as total_hire from employees group by month(hire_date) order by hire_month
```

```sql
--Find departments where average salary > 50,000.
select department,round(avg(salary))as avg_salary from employees group by department having avg(salary) > 50000
```

```sql
--departments where more than 5 employees earn > 60,000
select department,count(*) from employees where salary >60000 group by department having count(*)>5
```

```sql
--Find departments with fewer than 3 employees.
select department,count(*) from employees group by department having count(*)<3
```

```sql
--Find job titles with average salary > 70,000.
select job_title,round(avg(salary))as salary from employees group by job_title having avg(salary) > 70000
```

```sql
--Count employees per job title in each department.
select job_title,department,count(*)as employees from employees group by job_title,department
```

```sql
--Find total salary per city.
select city,sum(salary)as total_salary from employees group by city
```

```sql
--Find average salary per city.
select city,round(avg(salary)) as avg_salary from employees group by city 
```

```sql
--Find highest salary per department.
select department,max(salary) as high_salary from employees group by department
```

Advanced

```sql
--Find departments where the highest salary > 2 Ã— lowest salary.
select department,max(salary) as highest_salary ,min(salary) as lowest_salary from employees group by department having max(salary) > 2*min(salary )
```
