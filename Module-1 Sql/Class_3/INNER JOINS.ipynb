```sql
--list all the customers order with product details
-- INNER JOINS

select c.customer_id, 
c.first_name,c.last_name,o.order_id,o.product_name,o.quantity
from customers as  c inner join order_details as o
on c.customer_id=o.customer_id
```

```sql
--full breakdown like customers,orders,product_info
select 
od.order_id,c.first_name,od.product_name,od.quantity,od.price_each
from order_details od 
inner join orders o on od.order_id=o.order_id
inner join customers c on o.customer_id=c.customer_id
```

Easy Questions

```sql
--Show all orders with customer first name
select c.first_name,o.order_id
from customers c 
inner join orders o
on c.customer_id=o.customer_id
```

```sql
--Show customer full name and order date
select c.first_name,c.last_name,od.order_date
from customers c inner join order_details od
on c.customer_id=od.customer_id
```

```sql
--Show order_id and product_name
select o.order_id,od.product_name
from orders as o
inner join order_details as od
on o.order_id=od.order_id
```

```sql
--Show customer first name and product_name they ordered
select c.first_name,od.product_name
from customers c inner join order_details od
on c.customer_id=od.customer_id
order by c.first_name asc
```

Medium Level


```sql
--Show customer first name and total number of orders placed
select c.first_name,count(od.order_id) as totalNoOrder
from customers as c
inner join order_details as od
on c.customer_id=od.customer_id
group by c.first_name
```

```sql
--Show order_id and total quantity of items in each order
select o.order_id,sum(od.quantity) as total_quantity
from orders as o 
inner join order_details as od
on o.order_id=od.order_id
group by o.order_id
```

```sql
--Show customer name and total quantity they ordered
select c.first_name,od.quantity
from customers as c
inner join order_details as od
on c.customer_id=od.customer_id
order by c.first_name asc
```

```sql
--Display customers who ordered more than 5 items total
select c.first_name, sum(od.quantity) as total_items
from customers c
inner join orders o
on c.customer_id = o.customer_id
inner join order_details od
on o.order_id = od.order_id
group by c.first_name
having sum(od.quantity) > 5;
```

Hard Level

```sql
--Show top 3 customers who ordered highest quantity
select c.first_name,sum(od.quantity) as total_quantity
from customers as c
inner join orders as o
on c.customer_id=o.customer_id
inner join order_details as od
on o.order_id=od.order_id
group by c.first_name
order by sum(od.quantity) desc limit 3
```

```sql
--Show customer name and number of products in each order
SELECT 
    c.first_name AS customer_name,
    o.order_id,
    COUNT(od.product_id) AS num_of_products
FROM customers AS c
INNER JOIN orders AS o
    ON c.customer_id = o.customer_id
INNER JOIN order_details AS od
    ON o.order_id = od.order_id
GROUP BY c.first_name, o.order_id;
```

```sql
--Display orders where total quantity > 10
select c.first_name,sum(od.quantity) as total_quantity 
from customers as c
inner join orders as o
on c.customer_id=o.customer_id
inner join order_details as od
on o.order_id=od.order_id
group by c.first_name
having sum(od.quantity) > 10
```

```sql
--Show customer who made the most orders
select c.first_name,count(od.order_id) as most_order
from customers as c
inner join orders as o 
on c.customer_id=o.customer_id
inner join order_details as od
on o.order_id=od.order_id
group by c.first_name
order by count(od.order_id) desc limit 1
```
