```sql
--show inactive and active customers and their orders
select c.first_name,c.last_name,o.order_id,o.order_date

from customers as c
left join orders as o on c.customer_id=o.customer_id
```

```sql
--customers who never placed the orders
select 
c.first_name,c.last_name,o.order_id
from customers as c 
left join orders as o
on c.customer_id=o.customer_id
where o.order_id is null
```

Easy Level


```sql
--Show all customers and their orders (if any).
select c.first_name,o.order_id

from customers as c
left join orders as o 
on c.customer_id=o.customer_id
```

```sql
--Show all customers and the total number of orders they have made. If a customer has no orders, show 0.
select c.first_name,count(o.order_id)as total_orders
from customers as c
left join orders as o 
on c.customer_id=o.customer_id
group by c.first_name
```
