Level 1: Basic Joins


```sql
--Show customer first name and order ID.
select c.first_name,o.order_id

from customers c 
inner join orders o 
on c.customer_id=o.customer_id
```

```sql
--Show customer first name and total quantity of items they ordered.
select c.first_name,sum(od.order_id) as total_quantity

from customers c 
inner join order_details od
on c.customer_id=od.customer_id
group by c.first_name
```

```sql
--Display all orders along with the customer first and last names.
select c.first_name,c.last_name,o.order_id
from customers as c 
inner join orders as o 
on c.customer_id=o.customer_id
```

Level 2: Aggregations + Joins


```sql
--Show customer first name and total number of orders they placed.
select c.first_name,count(o.order_id)as total_order

from customers as c
inner join orders as o
on c.customer_id=o.customer_id
group by c.first_name
```

```sql
--Show customer first name and total amount spent, assuming price exists in order_details.
select c.first_name,sum(od.price_each) as total_spent

from customers as c
inner join order_details as od
on c.customer_id=od.customer_id
group by c.first_name
```

```sql
--Show customers who ordered more than 2 different products.
select c.first_name,count(distinct od.product_id) as different_products

from customers as c
inner join order_details as od
on c.customer_id=od.customer_id
group by c.first_name
having count(distinct od.product_id) > 2
```

Level 3: Filtering + Joins


```sql
--Show customer names and order IDs for orders with quantity > 5.
SELECT c.first_name, od.order_id, od.quantity
FROM customers AS c
INNER JOIN order_details AS od
ON c.customer_id = od.customer_id
WHERE od.quantity > 5;
```

```sql
--Show customers who placed orders in 2025 (or use order_date column).
select c.first_name,od.order_date

from customers as c
inner join order_details as od
on c.customer_id=od.customer_id
where od.order_date > '2025-01-01'
and od.order_date < '2026-01-01'
```

```sql
--Show orders along with customer name where the order total > 1000.
select c.first_name,sum(od.price_each * od.quantity) as total_order

from customers as c 
inner join order_details as od
on c.customer_id=od.customer_id
group by c.first_name
having sum(od.price_each*od.quantity) > 1000
```

Level 4: Multi-table Joins

```sql
--Show customer first name, order ID, and product name for all orders.
select c.first_name,od.order_id,od.product_name

from customers as c
inner join order_details as od
on c.customer_id=od.customer_id
```

```sql
--Show customer first name, product name, and quantity ordered.
select c.first_name,od.product_name,od.quantity as quantity_ordered

from customers as c
inner join order_details as od
on c.customer_id=od.customer_id
```

```sql
--Show customer first name, order ID, and category name (join with products → categories).
select c.first_name,od.order_id,od.category

from customers as c
inner join order_details as od
on c.customer_id=od.customer_id 
```

Level 5: Real-world Engineering Challenges


```sql
--Show top 5 customers by total order amount.
select c.first_name,sum(od.price_each*od.quantity)as total_amount 

from customers as c
inner join order_details as od
on c.customer_id=od.customer_id
group by c.first_name
order by total_amount desc limit 5
```

```sql
--Show all customers and their last order date.
select c.first_name,max(od.order_date) as last_order 

from customers as c
inner join order_details as od 
on c.customer_id=od.customer_id
group by c.first_name
```

Master 3-Table INNER JOIN Questions

```sql
--Show customer first name, order ID, and total order amount (quantity × price_each) per order.

select c.first_name,o.order_id,sum(od.quantity*od.price_each) as total_amount

from customers as c
inner join orders as o 
on c.customer_id=o.customer_id
inner join order_details as od
on o.order_id=od.order_id
group by c.first_name,o.order_id
```

```sql
--Show customer first name, order ID, and total quantity of items per order.
select c.first_name,o.order_id,sum(od.quantity) as total_quantity

from customers as c
inner join orders as o 
on c.customer_id=o.customer_id

inner join order_details as od
on o.order_id=od.order_id
group by c.first_name,o.order_id
```

```sql
--Show customer first name, last name, and number of distinct products in each order.
select c.first_name,c.last_name,od.order_id,count(distinct od.product_name) as distinct_products

from customers as c
inner join orders as o
on c.customer_id=o.customer_id

inner join order_details as od
on o.order_id=od.order_id
group by c.first_name,c.last_name,od.order_id
```

```sql
--Show top 5 customers by total quantity of products ordered across all their orders.
select c.first_name,sum(od.quantity) as total_quantity

from customers as c
inner join orders as o
on c.customer_id=o.customer_id
inner join order_details as od
on o.order_id=od.order_id
group by c.first_name
order by  sum(od.quantity) desc  limit 5
```
